- Dates 컬럼 데이터 타입은 문자열이므로 datetime 형으로 변환
- 필요한 년, 월, 일, 시, 분, 초 추출 후 새로운 컬럼으로 추가

```python
# datetime으로 변경하는 함수 정의
def to_datetime(data):
    data['Dates'] = pd.to_datetime(data['Dates'])
    
    data['year'] = data['Dates'].dt.year.astype(np.uint16)
    data['month'] = data['Dates'].dt.month.astype(np.uint8)
    data['day'] = data['Dates'].dt.day.astype(np.uint8)
    data['hour'] = data['Dates'].dt.hour.astype(np.uint8)
    data['minute'] = data['Dates'].dt.minute.astype(np.uint8)
    data['second'] = data['Dates'].dt.second.astype(np.uint8)
    
    return data
```

```python
# 날짜 데이터 datetimes64 타입으로 변경
train = to_datetime(train)
test = to_datetime(test)
```

```python
# datetime으로 변경됐는지 확인
train.dtypes
# test.dtypes
```

```python
# train 데이터셋 주말, 평일 추출 함수 정의
def weekday_weekend(data):
    # 방법1
    # data['weekday'] = [np.uint8(1) if (day == 'Saturday')|(day == 'Sunday') else np.uint(0) for day in data['DayOfWeek']]

    # 방법2
    # data['weekday'] = np.where((data['DayOfWeek'] == 'Saturday')|(data['DayOfWeek'] == 'Sunday'), 1, 0)

    # 방법3
    data['weekday'] = data['DayOfWeek'].apply(lambda x : 1 if x == 'Saturday' or x =='Sunday' else 0)
    data.head()
    
    return data
```

```python
# train 데이터셋 주말, 평일 추출후 컬럼 옆에 추가하기
train = weekday_weekend(train)
test = weekday_weekend(test)
```

```python
# 계절 컬럼 추가하는 함수 정의
def season(m):
  if 3 <= m <= 5:
    season = 'spring'
  elif 6 <= m <=8:
    season = 'summer'
  elif 9 <= m <= 11:
    season = 'fall'
  else:
    season = 'winter'

  return season
```

```python
# 계절 컬럼 추가하기
train['season'] = train['month'].apply(season)
test['season'] = test['month'].apply(season)
```

```python
# 계절 추가된지 확인하기
train['season'].unique()
```
